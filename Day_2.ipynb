```python
import pandas as pd
```

```python
passwords = pd.read_csv("passwords.csv")
passwords2 = pd.read_csv("passwords.csv")
passwords["count"] = 0
passwords["correct"] = False
passwords2["correct"] = False
```

```python
for row in range(len(passwords)):
    pword = passwords["password"][row]
    ltr = passwords["letter"][row]
    passwords["count"][row] = pword.count(ltr)
    if passwords["min_char"][row] <= passwords["count"][row] and passwords["count"][row] <= passwords["max_char"][row]:
        passwords["correct"][row] = True
        
passwords["correct"].value_counts()
```
