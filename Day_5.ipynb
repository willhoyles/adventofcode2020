```python
seats = open("seats.txt")
seat_ids = []
for line in seats:
    top = 127
    bottom = 0
    middle = top//2
    for i in range(6):
        print(line[i])
        if line[i] == "F":
            top = middle
        else:
            bottom = middle
        middle = bottom + ((top - bottom) // 2)
        print("top:"+str(top)+"middle:"+str(middle)+"bottom:"+str(bottom))
    if line[6] == "F":
        row = middle
    else:
        row = top
    print("row:"+str(row))
    up = 7
    down = 0
    half = up//2
    for i in range (7,10):
        print(line[i])
        if line[i] == "R":
            down = half
        else:
            up = half
        print("up:"+str(up)+"half:"+str(half)+"down:"+str(down))
        half = down + ((up - down) // 2)
    column = up
    print("column: "+str(column))
    ID = row * 8 + column
    seat_ids.append(ID)
    print(ID)
```

```python
print(max(seat_ids))
```

```python
seat_ids = sorted(seat_ids)
print(seat_ids)
```

```python
for i in seat_ids[1:]:
    if seat_ids[i] != (seat_ids[i-1] + 1):
        print(seat_ids[i] - 1)
```
